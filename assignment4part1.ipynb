```python
import tensorflow as tf
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
from tensorflow import keras

(train_images, train_labels), (test_images, test_labels) = keras.datasets.fashion_mnist.load_data()

scaled_train_images = train_images/255.0
scaled_test_images = test_images/255.0

val_set_size = math.ceil(scaled_train_images.shape[0]*.116)
scaled_val_images = scaled_train_images[:val_set_size]
val_labels = keras.utils.to_categorical(train_labels[:val_set_size], num_classes=10)
scaled_train_images = scaled_train_images[val_set_size:]
train_labels = keras.utils.to_categorical(train_labels[val_set_size:], num_classes=10)
encoded_test_labels = keras.utils.to_categorical(test_labels, num_classes=10)
print(f'scaled_val_images.shape: {scaled_val_images.shape}\nval_labels.shape: {val_labels.shape}\nscaled_train_images.shape: {scaled_train_images.shape}\ntrain_labels.shape: {train_labels.shape}\nscaled_test_images.shape: {scaled_test_images.shape}\nencoded_test_labels.shape: {encoded_test_labels.shape}')
```

```python
base_model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(64, (3,3), activation='relu', input_shape = (28,28,1)),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

base_model.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])

# Train the model (15 epo)
base_model_history = base_model.fit(scaled_train_images, train_labels, batch_size=32, epochs=30, verbose=1, validation_data=(scaled_val_images, val_labels))   

# scoring model
base_model_score = base_model.evaluate(scaled_test_images, encoded_test_labels, verbose=1)

# plotting loss curves
train_loss = base_model_history.history['loss']
val_loss = base_model_history.history['val_loss']
test_loss = base_model_score[0]

fig, ax = plt.subplots(1,3, figsize=(8,6))
ax[0].plot(range(1,31,1), train_loss, label='train_loss')
ax[0].plot(range(1,31,1), val_loss, label='val_loss')
ax[0].set_title('base model loss')
ax[0].set_xlabel('epoch')
ax[0].set_ylabel('loss')
ax[0].axhline(y=test_loss, color='red', label='test_loss')
plt.legend()
plt.tight_layout()

six_epoch_model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(64, (3,3), activation='relu', input_shape = (28,28,1)),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

six_epoch_model.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])

# Train the model
six_epoch_model_history = six_epoch_model.fit(scaled_train_images, train_labels, batch_size=32, epochs=6, verbose=1, validation_data=(scaled_val_images, val_labels))   

# scoring model
six_epoch_model_score = six_epoch_model.evaluate(scaled_test_images, encoded_test_labels, verbose=1)

# plotting loss curves
train_loss = six_epoch_model_history.history['loss']
val_loss = six_epoch_model_history.history['val_loss']
test_loss = six_epoch_model_score[0]

ax[1].plot(range(1,7,1), train_loss, label='train_loss')
ax[1].plot(range(1,7,1), val_loss, label='val_loss')
ax[1].axhline(y=test_loss, color='red', label='test_loss')
ax[1].set_title('6 epoch model loss')
ax[1].set_xlabel('epoch')
ax[1].set_ylabel('loss')
plt.legend()
plt.tight_layout()

filter_model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(64, (5, 5), activation='relu', input_shape = (28,28,1)),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64, (5,5), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

filter_model.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])

# Train the model
filter_model_history = filter_model.fit(scaled_train_images, train_labels, batch_size=32, epochs=6, verbose=1, validation_data=(scaled_val_images, val_labels))

# scoring model
filter_model_score = filter_model.evaluate(scaled_test_images, encoded_test_labels, verbose=1)

# plotting loss curves
train_loss = filter_model_history.history['loss']
val_loss = filter_model_history.history['val_loss']
test_loss = filter_model_score[0]

ax[2].plot(range(1,7,1), train_loss, label='train_loss')
ax[2].plot(range(1,7,1), val_loss, label='val_loss')
ax[2].axhline(y=test_loss, color='red', label='test_loss')
ax[2].set_title('6 epoch + (5,5) filter size')
ax[2].set_xlabel('epoch')
ax[2].set_ylabel('loss')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# getting predictions

base_model_predictions = np.argmax(base_model.predict(scaled_test_images), axis=1).tolist()
six_expoch_model_predictions = np.argmax(six_epoch_model.predict(scaled_test_images), axis=1).tolist()
filter_model_predictions = np.argmax(filter_model.predict(scaled_test_images), axis=1).tolist()

# getting actual labels

labels =['T-shirt/top','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle boot']

actual_labels = [labels[item] for item in test_labels]

base_model_predicted_labels = [labels[item] for item in base_model_predictions]
six_epoch_model_predicted_labels = [labels[item] for item in six_expoch_model_predictions]
filter_model_predicted_labels = [labels[item] for item in filter_model_predictions]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score
from matplotlib.ticker import PercentFormatter

# If this variable name is a typo in your notebook, fix it here:
six_epoch_model_predicted_labels = [labels[item] for item in six_expoch_model_predictions]
filter_model_predicted_labels = [labels[item] for item in filter_model_predictions]

# Put all models in one place
model_predictions = [
    ("Base Model", base_model_predicted_labels),
    ("Six-Epoch Model", six_epoch_model_predicted_labels),
    ("Filter Model", filter_model_predicted_labels),
]

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(24, 7), dpi=200)

for ax, (model_name, preds) in zip(axes, model_predictions):
    # Confusion matrix and accuracy
    cm = confusion_matrix(actual_labels, preds, labels=labels)
    acc = accuracy_score(actual_labels, preds)

    # Row-normalized matrix for color shading
    cm_sum = cm.sum(axis=1, keepdims=True)
    cm_norm = np.divide(cm, cm_sum, where=cm_sum != 0)

    # Heatmap
    im = ax.imshow(cm_norm, interpolation='nearest', cmap='Blues', vmin=0, vmax=1)

    # Axis formatting
    ax.set_xticks(np.arange(len(labels)))
    ax.set_yticks(np.arange(len(labels)))
    ax.set_xticklabels(labels, rotation=30, ha='right', fontsize=10)
    ax.set_yticklabels(labels, fontsize=10)
    ax.set_xlabel("Predicted label", fontsize=11)
    ax.set_ylabel("True label", fontsize=11)

    # Title
    ax.set_title(
        f"{model_name}\nAccuracy = {acc:.2%}",
        fontsize=15,
        fontweight='bold',
        pad=16
    )

    # White grid lines between cells
    ax.set_xticks(np.arange(cm.shape[1] + 1) - 0.5, minor=True)
    ax.set_yticks(np.arange(cm.shape[0] + 1) - 0.5, minor=True)
    ax.grid(which='minor', color='white', linestyle='-', linewidth=2)
    ax.tick_params(which='minor', bottom=False, left=False)

    # Annotate cells with count and row percentage
    threshold = cm_norm.max() / 2
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            count = cm[i, j]
            pct = cm_norm[i, j]
            text_color = 'white' if pct > threshold else 'black'

            ax.text(
                j, i,
                f"{count}\n",
                ha='center', va='center',
                color=text_color,
                fontsize=10,
                fontweight='bold'
            )

    # Remove outer spines for cleaner look
    for spine in ax.spines.values():
        spine.set_visible(False)

# Shared colorbar
# cbar = fig.colorbar(im, ax=axes, fraction=0.02, pad=0.02)
# cbar.ax.yaxis.set_major_formatter(PercentFormatter(xmax=1))
# cbar.set_label("Row-normalized percentage", fontsize=12)

# Overall title
fig.suptitle(
    "Confusion Matrix Comparison Across Models",
    fontsize=20,
    fontweight='bold',
    y=1.03
)

fig.tight_layout()

# Optional save
# plt.savefig("confusion_matrix_comparison.png", dpi=300, bbox_inches="tight")

plt.show()
```
